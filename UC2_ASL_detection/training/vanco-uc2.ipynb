```python
yaml_content = """
train: /kaggle/input/datasets/sanskar1104/dataset/dataset/train/images
val: /kaggle/input/datasets/sanskar1104/dataset/dataset/test/images
test: /kaggle/input/datasets/sanskar1104/dataset/dataset/test/images

nc: 10

names:
  0: A
  1: B
  2: C
  3: D
  4: E
  5: F
  6: G
  7: H
  8: I
  9: J
"""

with open("/kaggle/working/data_fixed.yaml", "w") as f:
    f.write(yaml_content)

print(open("/kaggle/working/data_fixed.yaml").read())
```

```python


from ultralytics import YOLO
import torch
import os

print("=" * 60)
print("PyTorch Version:", torch.__version__)
print("CUDA Available:", torch.cuda.is_available())
print("GPU Count:", torch.cuda.device_count())

for i in range(torch.cuda.device_count()):
    print(f"GPU {i}: {torch.cuda.get_device_name(i)}")

print("=" * 60)

DATA_YAML = "/kaggle/working/data_fixed.yaml"

model = YOLO("yolov8s.pt")

results = model.train(
    data=DATA_YAML,
    device=[0,1],

    epochs=150,
    imgsz=640,

    batch=16,
    workers=8,

    optimizer="AdamW",
    lr0=0.001,

    patience=25,

    save=True,
    save_period=1,

    cache=True,
    amp=True,

    pretrained=True,

    project="/kaggle/working/asl_detection",
    name="yolov8s",

    verbose=True
)

print("Training Complete")
```

```python
from ultralytics import YOLO

model = YOLO(
    "/kaggle/working/asl_detection/yolov8s-4/weights/best.pt"
)

metrics = model.val(
    data="/kaggle/working/data_fixed.yaml",
    plots=True,
    save_json=True
)

print(metrics)
```
